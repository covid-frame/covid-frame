```python
category_list = df.groupby(["category"]).size()
```

```python
reference_size = min(category_list.values)
reference_size
```

```python
list_idx = []
for idx, category in enumerate(category_list.keys()):
    index = df[df["category"]==category]
    list_idx.append(index)
```

```python
list_idx[0]
```

```python
from sklearn.utils import resample
```

```python
resampled_list = []
for idx in list_idx:
    resampled = resample(idx, n_samples=reference_size, replace=False, stratify=idx,
                         random_state=0)
    resampled_list.append(resampled)
```

```python
resampled_list[1]
```

```python
import pandas as pd
```

```python
n_df = pd.concat(resampled_list, ignore_index=True)
```

```python
n_df
```
